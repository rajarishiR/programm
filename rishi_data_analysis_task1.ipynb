```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
data1 = pd.read_csv("/content/college_1.csv")
data1.shape
```

```python
data2 = pd.read_csv("/content/college_2.csv")
data2.shape
```

```python
# merging those two dataframe into single dataframe
df = pd.concat([data1,data2])
df.shape
```

```python
df
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


```python
exceeded_expections = df.loc[df["CodeKata Score"]>15000]
exceeded_expections.to_csv("Exceeded expections")
exceeded_expections
```

###### if  10000<codekata score<15000   (Reached_expectations.csv)



```python
Reached_expections = df.loc[df["CodeKata Score"].between(10000,15000)]
Reached_expections.to_csv("reached_expections")
Reached_expections
```

###### if  7000<codekata score<10000   (Needs_Improvement.csv)


```python
Needs_improvement = df.loc[df["CodeKata Score"].between(7000,10000)]
Needs_improvement.to_csv("needs_improvement")
Needs_improvement
```

###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
unsatifactory = df.loc[df["CodeKata Score"]<7000]
unsatifactory.to_csv("unsatifactory")
unsatifactory
```
